```python
def plotHistogram(img):
    g1 = 1.79526731311
    g2 = 1.85695064051
    g3 = 1.69743757856
    rChannel = img[:,:,0]
    gChannel = img[:,:,1]
    bChannel = img[:,:,2]
    rChannel = np.power(rChannel, g1)
    gChannel = np.power(gChannel, g2)
    bChannel = np.power(bChannel, g3)
    rChannel = np.dstack((rChannel, gChannel))
    img = np.dstack((rChannel, bChannel))
    img = img.astype(np.uint8)
    histr = cv2.calcHist([img],[0],None,[255],[0,256])
    plt.figure()
    plt.subplot(1,3,1)
    plt.plot(histr,color = 'r')
    plt.xlim([0,255])
    
    histr = cv2.calcHist([img],[1],None,[25],[0,256])
    plt.subplot(1,3,2)
    plt.plot(histr,color = 'g')
    plt.xlim([0,25])
    
    histr = cv2.calcHist([img],[2],None,[25],[0,256])
    plt.subplot(1,3,3)
    plt.plot(histr,color = 'b')
    plt.xlim([0,25])
    
```
